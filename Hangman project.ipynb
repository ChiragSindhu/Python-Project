```python
import random

# Functions
def validation(userInput):
    if(len(userInput) == 0):
        return "One letter Required!!"
    
    if(len(userInput) > 1):
        return "Only one letter allowed!!"
    
    asciiVal = ord(userInput)
    isValid = -1
    if(asciiVal  >= 65 and asciiVal <= 90):
        isValid = 1
    if(asciiVal  >= 97 and asciiVal <= 122):
        isValid = 1
    if(isValid == -1):
        return "Invalid Input"
    return 0

# Printing nice text
def write(color_code,text):      
    if(color_code == "red black"):
        print("\x1b[4;1;37;40m" + text + "\x1b[0m")
    elif(color_code == "gray"):
        print("\x1b[0;1;30;47m" + text + "\x1b[0m")
    elif(color_code == "green"):
        print("\x1b[4;1;32;46m" + text + "\x1b[0m")
    elif(color_code == "red"):
        print("\x1b[1;31m" + text + "\x1b[0m")
    elif(color_code == "bold"):
        print("\033[1m" + text + "\033[0m")
    else:
        print(text)

guessList = {
    'homework' : 'Teachers give this',
    'shoes' : 'Item of clothing',
    'spider' :  'Insect with 8 legs',
    'drums' : 'Musical instrument you hit',
    'dentist' : 'Takes care of teeth',
    'potato' : 'Common food, grown underground',
    'lays' : 'Popular snack'
}

word = random.choice(list(guessList.items()))
hint = word[1]
word = word[0]

maxGuess = 4
currentWord = ""

while(maxGuess):
    # print(word + " : " + currentWord)
    write("red black","////////////////////// - HANG MAN - ////////////////////////////////")
    if(len(word) == len(currentWord)):
        write("green","You guessed the word \"" + word  + "\" correctly")
        write("green","YOU WON!")
        break
        
    write("gray"," > Hint : \"" + hint + "\"")
    
    userGuesses = ""
    for i in range(len(word)):
        if word[i] in currentWord:
            userGuesses = userGuesses + word[i].upper()
        else:
            userGuesses = userGuesses + " _ "
    write("bold","Word => " + userGuesses)
    
    write("bold","Remaining Guesses  : " + str(maxGuess))
    maxGuess = maxGuess -  1
    
    userInput = input("Your Guess (only one letter) : ");
    
    validData = validation(userInput)
    if(validData != 0):
        maxGuess = maxGuess + 1
        write("red"," > Error :\"" + validData  + "\"")
        continue
    
    userInput = userInput.lower()
    
    if userInput in currentWord:
        write("red black","Word Already Done! try new one")
        maxGuess = maxGuess + 1
        continue
    else:        
        if userInput in word:
            write("green",random.choice(["Well Done","Keep Going","Perfect!","Very Good"]));
            maxGuess = maxGuess + 1
        else:
            if(maxGuess != 0):
                write("red black","Oops! try again")
            else:
                write("red black","Game Over!")
                
        for i in range(len(word)):
            if(word[i] == userInput):
                currentWord = currentWord + userInput;
    
    print(" ")
```
